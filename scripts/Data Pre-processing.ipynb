# Data Pre-processing

**Import datasets into python**

```python
import pandas as pd
match_df = pd.read_csv(r"PATH TO AGG DATASET")
match_df
```

```python
kill_df = pd.read_csv(r"PATH TO KILL DATASET")
kill_df
```

**1.Data Reduction (Sampling - selecting a random subset of the original data for analysis):** 
    <br/>
>Delete rows while keeping only the first 10000 rows for both datasets to use

```python
match_df.drop(match_df.index[10000:], inplace=True)
match_df
```

```python
kill_df.drop(kill_df.index[10000:], inplace=True)
kill_df
```

**2.Change the victim names in victim_name to player1, player2,...**

```python
print(kill_df['victim_name'])
```

```python
y=0
x=1

for _ in kill_df['victim_name']:
    kill_df.loc[y, 'victim_name'] = (f"Player {x}")
    x = x+1
    y = y+1

print(kill_df['victim_name'])
```

**3.Data Reduction (Feature selection - selecting a subset of the original features that are most relevant to the analysis):**
    <br/>
>Delete map 'MIRAMAR' while keeping map 'ERANGEL' in kill_df Display the column "map" consisting of both MIRAMAR and ERANGEL maps

```python
print(kill_df['map'])
```

```python
kill_df = kill_df[kill_df.map != 'MIRAMAR']
print(kill_df['map'])
```

**4.Data cleaning (Smooth Noisy Data - removing noise from a data set, allowing important patterns to stand out):**
<br/>
>Clean the date to ie. 2017-11-20

```python
print(match_df['date'])
```

```python
match_df['date'] = match_df['date'].astype(str).str[:10]
print(match_df['date'])
```

**5.Data Transformation (Data normalisation - standardizing the data to a common scale or range):**
<br/>
>Convert 'player_survive_time' and 'time' in seconds to time in minutes, ie. 75 seconds = 1.25 min

```python
print(match_df['player_survive_time'])
```

```python
match_df['player_survive_time'] = match_df['player_survive_time'] /60
print(match_df['player_survive_time'])
```

```python
print(kill_df['time'])
```

```python
kill_df['time'] = kill_df['time'] /60
print(kill_df['time'])
```
